```python
dfs = {}
decomposed_dfs = {}
def decompose(home):
    dfs[home] = store['/%d' %home]['2013']['use'].resample("15T").fillna(method="ffill")
    res = sm.tsa.seasonal_decompose(dfs[home].values, freq=96*7) 
    return pd.DataFrame({"seasonal":res.seasonal, "trend":res.trend, "obs":res.observed}, index=dfs[home].index)
    
```
