```python
def chunker(seq, size):
    return (seq[pos:pos + size] for pos in xrange(0, len(seq), size))
```

```python
nprocs = 4

pool = mp.Pool(nprocs)

def myfunction(df):
    return df.mean(), df.shape

for chunk in chunker(df, nprocs):
    print chunk, type(chunk)
    data = pool.map(myfunction, chunk)
    
```
